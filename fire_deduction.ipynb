```python
import cv2
import random
from ultralytics import YOLO
import winsound  # For alarm sound on Windows

# Load the YOLO model
model = YOLO(r"D:\AI COURSE TAMIL\First project\fire.v2i.yolov8\runs\detect\train2\weights\best.pt")

# Define a dictionary to store colors for each class
colors = {}

# Define the color red
red_color = (0, 0, 255)  # OpenCV uses BGR format

# Open the webcam
cap = cv2.VideoCapture(0)

# Get screen dimensions
screen_width = 1920  # Replace with your screen width
screen_height = 1080  # Replace with your screen height

# Set window dimensions to 80% of the screen size
window_width = int(screen_width * 0.8)
window_height = int(screen_height * 0.8)

# Create a named window and set it to normal
cv2.namedWindow("Webcam", cv2.WINDOW_NORMAL)
cv2.resizeWindow("Webcam", window_width, window_height)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Perform prediction
    results = model(frame)

    # Initialize a flag to check for detection
    detected = False

    # Process the results
    for result in results:
        boxes = result.boxes
        for box in boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            conf = box.conf[0]
            cls = int(box.cls[0])

            # Set color to red if class is 'fire'
            if model.names[cls] == 'fire':
                color = red_color
                detected = True
            else:
                # Assign a color to the class if not already assigned
                if cls not in colors:
                    colors[cls] = [random.randint(0, 255) for _ in range(3)]
                color = colors[cls]

            # Draw rectangle around the detected object
            cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)

            # Put label above the rectangle
            label = f'{model.names[cls]} {conf:.2f}'
            cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, color, 2)

    # Trigger the alarm if fire is detected
    if detected:
        # Sequence of beeps
        winsound.Beep(1000, 500)  # Beep at 1000 Hz for 500 milliseconds
        winsound.Beep(1500, 500)  # Beep at 1500 Hz for 500 milliseconds

    # Display the frame
    cv2.imshow("Webcam", frame)

    # Break the loop on 'q' key press
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the capture and close the windows
cap.release()
cv2.destroyAllWindows()
```
